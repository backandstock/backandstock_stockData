## 키움증권 API를 이용한 금융데이터 수집

```python
from pykiwoom.kiwoom import *
import sys 
from PyQt5.QtWidgets import *
from PyQt5.QAxContainer import *
import pythoncom
import pandas as pd 
import numpy as np
```

### 로그인

```python
kiwoom = Kiwoom()
kiwoom.CommConnect(block=True)
print("블록킹 로그인 완료")
```

### 주식정보 수집

```python
# 시장별 주식 코드 가져오기 (list에 주식 코드만 담겨서 반환됨)
kospi = kiwoom.GetCodeListByMarket('0')
kosdaq = kiwoom.GetCodeListByMarket('10')
etf = kiwoom.GetCodeListByMarket('8')
```

```python
# 주식 정보 저장을 위한 빈 데이터프레임
stockInfo = pd.DataFrame(columns = {'stockName','stockCode','publicDate', 'market'})
```

```python
# 시장별 dict 생성
codeDict = {'kospi':kospi, 'kosdaq':kosdaq, 'etf':etf}

# 시장별로 데이터 가져오기 
for key in list(codeDict.keys()):
    for targetCode in codeDict[key]:
        name = kiwoom.GetMasterCodeName(targetCode)
        publicDate = kiwoom.GetMasterListedStockDate(targetCode)
        addingData = {"stockName":name, "stockCode" : targetCode, 
                      "publicDate":publicDate.strftime("%Y%m%d"), "market":key}
        stockInfo = stockInfo.append(addingData , ignore_index=True)
```

```python
# 중복된 코드 정보는 삭제함 (같은 종목 코드가 etf, kospi 각각 들어가 있는 경우가 있음)
stockInfo.drop_duplicates(["stockCode"], inplace = True)
```

### 주식월봉차트조회 요청

```python
# 차트 조회를 위한 주식 코드 리스트 생성
codeList = list(stockInfo['stockCode'])

# 문자열로 오늘 날짜 얻기
now = datetime.datetime.now()
today = now.strftime("%Y%m%d")

# 월봉 차트 데이터 저장을 위한 빈 데이터 프레임 생성
stockData = pd.DataFrame( columns = {'현재가', '거래량', '거래대금', '일자', '시가', '고가', '저가', 
                                     '수정주가구분', '수정비율', '대업종구분',
                                     '소업종구분', '종목정보', '수정주가이벤트', '전일종가'})

# 전 종목의 월봉
for i, code in enumerate(codeList[176:]):
    print(f"{i}/{len(codeList)} {code}")
    df = kiwoom.block_request("opt10083",
                              종목코드=code,
                              기준일자=today,
                              수정주가구분=1,
                              output = "주식월봉차트조회",
                              next=0)
    df['code'] = code
    stockData = pd.concat([stockData, df], axis = 0)
    # api 요청 차단 대비를 위해 time.sleep을 추가 
    time.sleep(3.6)
```

```python
# 데이터 백업하기
stockInfo.to_csv('stockInfo.csv')
stockData.to_csv('stockData.csv')
```
